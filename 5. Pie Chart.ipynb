```python
# Pie Chart using matplotlib
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%.1f%%",shadow=True)
# autopct is used for showing chart into the percentage
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%.1f%%",explode=[0,0.1,0,0,0])

# explode is used for the exploding the graph
```
