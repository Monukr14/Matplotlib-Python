```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,'g+--',linewidth=2)
# Here we are doing format specifier of the graph ("g+--" this is format specifier)
# visual properities by passing argument   , here "g+" means g is green color and + is marker.
# It should be always in the 3rd argument
```

```python
plt.plot(x,y,color="blue",marker="D",linestyle="-.")
```

```python
plt.plot(x,y,color="blue",marker="D")
```

```python
plt.plot(x,y,color="blue",marker="D",markersize=10) 
# marksize is used for increasing the size of marker here marker is D it means Diamond
```

```python
plt.plot(x,y,color="blue",alpha=0.2) 
# alpha is used as opacity , it will effect as transparent
```
